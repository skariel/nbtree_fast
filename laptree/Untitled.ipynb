```julia
include("./lib/all.jl")
```

```julia
particles = make_particles(1000000);
ph = get_parent_hierarchy(particles);
```

```julia
plot([p._x for p in particles], [p._y for p in particles], "k.", ms=0.01)
l = get_leafs(ph[1:end-11], 1)
plot([p._x for p in l], [p._y for p in l], "r.",ms=0.3)
```

```julia
errs,ixs = get_err_arr(ph,particles,1000,0.5,0.001)
```

```julia
median(abs(errs))
```

```julia
eps2 = 0.001
alpha2 = 0.5
N = 1000
errs,ixs = get_radial_err_arr(ph, particles, N, alpha2, eps2)
```

```julia
median(errs)
```

```julia
x=[p._x for p in particles][ixs]
y=[p._x for p in particles][ixs]
z=[p._x for p in particles][ixs]
r = sqrt(x.*x+y.*y+z.*z);
```

```julia
plot(r,errs,"r.",ms=0.9)
```
