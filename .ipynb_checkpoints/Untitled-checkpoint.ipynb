```julia
using PyPlot
```

```julia
function calc_f_cm(x,y,z,rx,ry,rz)
    cmx = mean(x)
    cmy = mean(y)
    cmz = mean(z)
    dx = cmx-rx
    dy = cmy-ry
    dz = cmz-rz
    dr2 = dx.*dx + dy.*dy + dz.*dz
    dr3 = dr2.*sqrt(dr2)
    length(x).*dx./dr3
end

function calc_f_ex(x,y,z,rx,ry,rz)
    f = zeros(length(rx))
    for i in eachindex(x)
        dx = x[i]-rx
        dy = y[i]-ry
        dz= z[i]-rz
        dr2 = dx.*dx + dy.*dy + dz.*dz
        dr3 = dr2.*sqrt(dr2)
        f += dx./dr3
    end;
    f
end

function calc_err(x,y,z,rx,ry,rz)
    f_cm = calc_f_cm(x,y,z,rx,ry,rz)
    f = calc_f_ex(x,y,z,rx,ry,rz)
    (f_cm-f)./f
end;
```

```julia
rx = -5.0
ry = 0.0
rz = 0.0

NIT = 1000
NLN = 24
errs = zeros((NIT,NLN))
for it in 1:NIT
    ix = 0
    for lN in linspace(1,6,NLN)
        ix += 1
        N = round(Int64, 10.^lN)
        x = rand(N)-0.5
        y = rand(N)-0.5;
        z = rand(N)-0.5
        errs[it,ix] = calc_err(x,y,z,rx,ry,rz)[1]
    end
end;
```

```julia
for i in 1:NIT
    plot(collect(2:NLN)+rand()*0.03,errs[i,2:end],"k.", ms=0.1)
end
```

```julia
me = [median(errs[:,i]) for i in 1:NLN];
plot(me)
```

```julia
cm_err = calc_err(x,y,z,rx,ry,rz)

plot(rx,cm_err,"-b",ms=0.9)

grid(true)
xlabel("r [distance]")
ylabel("log(err)")
```
